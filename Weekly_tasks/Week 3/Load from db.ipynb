```python
import sqlite3
import pandas as pd
```

```python
def db_to_csv(db_path, table_name, output_file_name):
    conn = sqlite3.connect(db_path)
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    df.to_csv(f"{output_file_name}.csv", index=False)
    conn.close()
```

```python
db_to_csv(r'\Weekly_tasks\Week 2\olxDB.db', "listing", "listing")
db_to_csv(r'\Weekly_tasks\Week 2\olxDB.db', "photos", "photos")
```
