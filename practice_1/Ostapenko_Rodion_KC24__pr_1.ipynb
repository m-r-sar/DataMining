### I. Numerical Computing with NumPy.
#### 1. Array Creation & Basic Operations

```python
#A. Task: Create a 1D array of 10 random integers. Create a 3x3 identity matrix.
arr = np.random.randint(10, 100, 10)
print(arr)
identity_matrix = np.identity(3)
print(identity_matrix)
```

```python
#B. Task: Multiply the 1D array by a scalar value without using any for loops.
print(arr*10)
```

```python
#C. Task: Create two 3x3 matrices. Perform both element-wise multiplication and the dot product.
matrix_1 = np.array([[1, 2, 3],
            [3, 2, 1],
            [7, 6, 9]])

matrix_2 = np.array([[2, 2, 2],
            [3, 3, 3],
            [10, 10, 10]])

print(matrix_1*matrix_2)
print(np.dot(matrix_1, matrix_2))
```

#### 2. Statistical Methods

```python
#A. Task: Generate an array of 100 random numbers from a normal distribution.
arr = np.random.randn(100)
arr
```

```python
# B. Task: Calculate and print the mean, median, standard deviation, and variance of the array. Find the index of the maximum value.
mean = np.mean(arr)
median = np.median(arr)
standard_deviation = np.std(arr)
variance = np.var(arr)
max_index = np.argmax(arr)
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {standard_deviation}")
print(f"Variance: {variance}")
print(f"Max Index: {max_index}")
```

### II. Tabular Data with Pandas
#### 1. Series


```python
#A. Task: Create a Series object
series = pd.Series([0, -1, 4, -3, 2])
print(series)
```

```python
#B. Task: Modify data in the Series (only single value, and all values)
series[1] *= -1
print(series)

series += 10
print(series)
```

```python
#C. Task: Modify index of the Series
series.index = [f"Index: {-i}" for i in range(len(series))]
print(series)
```

#### 2. DataFrames

```python
#A. Task: Create a Pandas DataFrame from a Python dictionary.
dict = {
    "A": np.random.randint(1, 100, 10), 
    "T": np.random.randint(1, 100, 10), 
    "C": np.random.randint(1, 100, 10), 
    "G": np.random.randint(1, 100, 10)
    }

dataframe = pd.DataFrame(dict, index=[f"Sequence: {i+1}" for i in range(10)])
dataframe
```

```python
#B. Task: Construct a DataFrame from a 2D NumPy array, manually assigning column names (e.g., 'Feature_1', 'Feature_2', 'Target').
array_2d = np.random.randn(3, 4)
df = pd.DataFrame(array_2d, columns = ["Numbers A", "Numbers B", "Numbers C", "Numbers D"])
df
```

```python
# C. Task: Create own column as sum (or any other result) of using other columns
dataframe["Total length"] = [dataframe.values[i].sum() for i in range(10)]
dataframe
```

#### 3. Data Selection and Indexing

```python
#A. Task: Use .loc to select rows based on a specific index label.
dataframe.loc["Sequence: 4"]
```

```python
#B. Task: Use .iloc to select the first 5 rows and the last 2 columns of the DataFrame.
dataframe.iloc[:5, 3:]
```

```python
#C. Task: Filter the DataFrame to show only rows where 'Feature_1' is greater than a specific threshold.
dataframe[(dataframe["A"]>50) & (dataframe["G"]<60)]
```
