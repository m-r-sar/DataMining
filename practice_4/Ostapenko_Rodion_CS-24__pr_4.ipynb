### I. Working with Datetime

```python
df = pd.read_csv(r"src\hourlySteps_merged.csv")
df.dtypes
```

```python
df.head(10)
```

```python
#Convert the string representation of the date/time into a proper pandas datetime object using pd.to_datetime().
df['ActivityHour_datetime'] = pd.to_datetime(df['ActivityHour'], format='%m/%d/%Y %I:%M:%S %p')
df.head(10)
```

```python
df.dtypes
```

```python
#Extract at least three new useful features from the datetime column (e.g., year, month, day of the week, hour, or an is_weekend boolean flag).
df["Hour"] = df["ActivityHour_datetime"].dt.hour
df["Day"] = df["ActivityHour_datetime"].dt.day
df["Month"] = df["ActivityHour_datetime"].dt.month
df["Year"] = df["ActivityHour_datetime"].dt.year

df["Day_of_Week"] = df["ActivityHour_datetime"].dt.day_name()
df["Is_Weekend"] = df["ActivityHour_datetime"].dt.weekday > 4

df.head(10)
```

```python
df.dtypes
```

```python
#Drop the original raw datetime string column from the working DataFrame.
df = df.drop(columns = ['ActivityHour'])
df.head(10)
```

### III. Working with other string data

```python
#Identify a column containing unstructured text, mixed formats, or string artifacts (e.g., "$1,000", " Category A ", "ID-4920").
string_data_df = df[['menu_item_name', 'discount_code']]
string_data_df.head(10)
```

```python
#Perform basic string operations using the .str accessor (e.g., lowercasing, stripping trailing/leading whitespaces, replacing specific characters).
string_data_df['menu_item_name'] = string_data_df['menu_item_name'].str.lower().str.replace(' ', '_').str.strip()
string_data_df.head(10)
```

```python
nonull_string_data_df = df[['menu_item_name', 'discount_code']]
nonull_string_data_df = nonull_string_data_df.dropna(axis=0, how='any')
nonull_string_data_df.head(10)
```

```python
#Extract a specific pattern using Regular Expressions (Regex) via pandas string methods and save the extracted data as a new feature column.
nonull_string_data_df['promo_id'] = nonull_string_data_df['discount_code'].str.extract(r'PROMO-(.*)')
nonull_string_data_df.head(10)
```

### IV. Balanced encoding

```python
#Identify a categorical column with high cardinality (many unique categories), show cardinality statistics by categorical columns.

cols = df.select_dtypes(include=['str']).columns
for col in cols:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
#menu_item_name column has a high cardinality
#Apply a balanced/advanced encoding technique (such as Target Encoding, Frequency/Count Encoding, or Hashing) to this column.

#Frequency Encoding
encoding_df = df[['menu_item_name', 'menu_item_id', 'total_amount']]
freq_map = encoding_df['menu_item_name'].value_counts()
encoding_df['menu_item_freq'] = encoding_df['menu_item_name'].map(freq_map)
encoding_df.head(10)
```

```python
#Target Encoding
target_mean = encoding_df.groupby('menu_item_name')['total_amount'].mean()

encoding_df['menu_item_target_enc'] = encoding_df['menu_item_name'].map(target_mean)

encoding_df.head(10)
```

#### Observe how this affects the dataset compared to standard One-Hot Encoding.
- Original Columns: 1

- One-Hot Encoding would create: 1034 columns (one column for each unique)

- Frequency Encoding created: 1 column

- Target Encoding created: 1 column
