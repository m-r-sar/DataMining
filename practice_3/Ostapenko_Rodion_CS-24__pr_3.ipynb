### I. Prepare the data with missing values.


```python
def make_missing_values(
    src_path: str,
    tgt_path: str,
    n_rows: int = 100,
    missing_per_col: float = 0.1,
) -> None:
    """
    This function loads your DataFrame and stores it as a separate .csv file for your experiments with missing data.
    
    :param src_path: Path to the original source .csv file to read.
    :param tgt_path: Path where the modified .csv file with missing values will be saved.
    :param n_rows: Number of rows to read from the source file.
    :param missing_per_col: The fraction of values to drop per column (e.g., 0.2 for 20% missing data). Must be between 0.0 and 1.0.
    :return: None
    """
    
    # 1. Load the specific number of rows from the source CSV
    df = pd.read_csv(src_path, nrows=n_rows)
    
    # 2. Calculate the exact number of missing values to inject per column
    # We use len(df) in case the source file has fewer rows than n_rows
    n_missing = int(len(df) * missing_per_col)
    
    # 3. Iterate through each column and inject NaNs randomly
    for col in df.columns:
        # Select random indices without replacement, so we don't pick the same row twice
        missing_indices = np.random.choice(df.index, size=n_missing, replace=False)
        
        # Assign NaN to the selected random indices for the current column
        df.loc[missing_indices, col] = np.nan
        
    # 4. Save the modified DataFrame to the target path
    df.to_csv(tgt_path, index=False)
```

```python
make_missing_values(
    src_path=r".\src\boston.csv",
    tgt_path=r".\miss_boston.csv",
)
```

### II. Working with missing values


```python
#1. Remove rows with any missing values (Do NOT same results in the working DataFrame).
df_remove_rows = df_miss.dropna(axis=0, how='any')
df_remove_rows.head(10)
```

```python
df_remove_rows.isnull().sum()
```

#### This method is best used when the dataset is very large, and the rows with missing data represent only a tiny fraction of the total data. But it is dangerous if it causes a massive loss of usable data.

```python
#2. Remove cols with any missing values (Do NOT same results in the working DataFrame).
df_remove_cols = df_miss.dropna(axis=1, how='any')
df_remove_cols.head(10)
```

#### This is a drastic measure, okay to use only when a specific column is missing a massive percentage of its data. Otherwise, you risk throwing away highly valuable features

```python
#3. Remove rows with the specific number (or percent) of missing values (Do NOT same results in the working DataFrame).
df_remove_rows_by_number = df_miss.dropna(axis=0, thresh=10)
df_remove_rows_by_number.head(10)
```

```python
df_remove_rows_by_number.isnull().sum()
```

#### This is a much smarter alternative to dropping rows blindly. It allows us to preserve rows that have mostly good, usable data, while safely discarding rows that are practically blank.

```python
#4. Execute back filling of missing values (Do NOT same results in the working DataFrame).
df_bf = df_miss.bfill()
df_bf.head(10)
```

```python
df_bf.isnull().sum()
```

```python
#5. Execute front filling of missing values (Do NOT same results in the working DataFrame).
df_ff = df_miss.ffill()
df_ff.head(10)
```

```python
df_ff.isnull().sum()    
```

```python
#6. Execute filling by a specific values (median, mean, just value indicator like -1 for numerical or "missing" for object).

df_mean_fill = df_miss.fillna(df_miss.mean())
df_mean_fill.head(10)
```

```python
df_miss["TAX"].describe()
```

```python
df_mean_fill["TAX"].describe()
```

#### The overall mean of the column naturally remained exactly the same. However, the standard deviation decreased. This happens because we added multiple new data points exactly at the center of the distribution, which reduces the variance of the data points from the center

#### Best used for continuous, normally distributed numerical data that does not have extreme outliers

```python
df_median_fill = df_miss.fillna(df_miss.median())
df_median_fill.head(10)
```

```python
df_miss["TAX"].describe()
```

```python
df_median_fill["TAX"].describe()
```

#### The mean shifted slightly closer to the median, and the standard deviation also decreased, though often less uniformly than mean-filling. This method safely preserved the general shape of our skewed data without letting outliers dictate the missing values.

```python
df_indicator_fill = df_miss.fillna(-1)
df_indicator_fill.head(10)
```

#### This is excellent when the absence of data is actually a signal itself like for creating a cluster. But for my dataset can act weird on statistics

### Final cleaning

```python
df_clean = df_miss.dropna(axis=0, thresh=11)
df_clean["CHAS"] = df_clean["CHAS"].fillna(0)
df_clean = df_clean.fillna(df_clean.median())

df_clean.head(10)
```

```python
#final proof
df_clean.isnull().sum().sum() == 0
```

#### The dataset has been successfully prepared, with missing values handled based on their data types and distributions
